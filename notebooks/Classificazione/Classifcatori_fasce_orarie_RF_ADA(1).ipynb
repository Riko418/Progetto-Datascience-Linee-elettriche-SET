```python
def print_evaluation_scores(y_train, y_train_pred, y_test, y_test_pred, is_test=True):
    if is_test:
        train_dataset_name = "Train+Valid"
        test_dataset_name = "Test"
        train_y_true = y_train
        train_y_pred = y_train_pred
        test_y_true = y_test
        test_y_pred = y_test_pred
    else:
        train_dataset_name = "Train"
        test_dataset_name = "Valid"
        train_y_true = y_train
        train_y_pred = y_train_pred
        test_y_true = y_test
        test_y_pred = y_test_pred

    train_accuracy = accuracy_score(train_y_true, train_y_pred)
    train_f1 = f1_score(train_y_true, train_y_pred)
    train_precision = precision_score(train_y_true, train_y_pred)
    train_recall = recall_score(train_y_true, train_y_pred)

    test_accuracy = accuracy_score(test_y_true, test_y_pred)
    test_f1 = f1_score(test_y_true, test_y_pred)
    test_precision = precision_score(test_y_true, test_y_pred)
    test_recall = recall_score(test_y_true, test_y_pred)

    print(f"Accuracy per {test_dataset_name}: {test_accuracy} e Accuracy per {train_dataset_name}: {train_accuracy}")
    print(f"F1 score per {test_dataset_name}: {test_f1} e F1 score per {train_dataset_name}: {train_f1}")
    print(f"Precision per {test_dataset_name}: {test_precision} e Precisione per {train_dataset_name}: {train_precision}")
    print(f"Recall per {test_dataset_name}: {test_recall} e Recall per {train_dataset_name}: {train_recall}")
```

Tramite la funzione appena definita valuto le predizioni fatte per la fascia diurna e notturna e rappresento le confusion matrix sia del train che dell'insieme di validazione per avere una rappresentazione più grafica di cosa ottengo veramente e con quale tipo di errore

```python
def plot_confusion_matrix(y_train, y_train_pred, y_test, y_test_pred, classes, is_test=True):
    #OUTPUT: due confusion matrix una affianco all'altra, una per il test e l'altra per il train
    
    #ANNOTAZIONI: classes sono le classi del regressore che si utilizza, si possono passare come 'nome_regressore.classes_'
    #'is_test' può essere solo True o False, mi serve solo per dare titoli diversi alle due matrici di confusione, ovviamente 
    #se è True la y_train sarà in realtà la y del train + y_valid, mentre se è False allora y_test in realtà è la y_valid
    #proprio per questo cambio i nomi ai titoli delle confusion matrix
    
    fig, axs = plt.subplots(1, 2, figsize=(10, 5))
    if is_test:
        titles = ["Confusion Matrix Train+Valid", "Confusion Matrix Test"]
    else:
        titles = ["Confusion Matrix Train", "Confusion Matrix Validazione"]
    
    for i, (y_values, y_pred) in enumerate([(y_train, y_train_pred), (y_test, y_test_pred)]):
        cm = confusion_matrix(y_values, y_pred)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classes)
        disp.plot(ax=axs[i], cmap='Blues')
        axs[i].set_title(titles[i])
        axs[i].set_xlabel('Predicted Labels')
        axs[i].set_ylabel('True Labels')
    
    plt.tight_layout()
    plt.show()
```

Noto che faccio ENORME OVERFITTING, quindi cerco i valori migliori dei quattro iperparametri fatti in classe e di n_estimators. Inoltre si nota come i dati della fascia diurna risultino classificati in maniera meno ottimale rispetto a quelli della fascia notturna dove sì, si nota overfitting, ma comunque i risultati sull'insieme di validazione sono praticamente ottimali.

```python
def train_classification_model(classificator, param_grid, cv, X, y):
    
    #OUTPUT: regressore ottimizzato con i parametri messi nella griglia di parametri param_grid
    
    #ANNOTAZIONI: X e y sono le X e le y del TRAIN perchè questa funzione mi fitta il regressore che voglio usare con i migliori
    #iperparametri che ho messo nella griglia di parametri param_grid.
    #'param_grid' è la griglia di parametri che deve essere definita come un dizionario quindi:
    #param_grid={
    #            "parametro_1": "[valore1, valore2, ...]",
    #            "parametro_2": "[valore1, valore2]",
    #             ...
    #            }
    #La funzione in outpunt mi restituisce il regressore ottimizzato quindi la devo chiamare così:
    # nome_regressore_ottimizzato = train_regression_model(regressore_da_ottimizzare, param_grid, ...)
    #Nota: il regressore migliore 'best_regressor' NON fitta sui punti, non è stata implementata quella riga di codice perchè
    #se i 'best_params' sono gli stessi tra un tentativo e l'altro quando cambio i parametri di param_grid non ha senso aspettaere
    #che mi faccia il fit, è lo stesso di prima già fatto
    
    
    grid_search = GridSearchCV(classificator, param_grid, cv=cv)
    grid_search.fit(X, y)

    #Printa i migliori parametri trovati
    best_params = grid_search.best_params_
    print("Migliori parametri:", best_params)

    
    best_classificator = classificator.__class__(**best_params) #Crea un nuovo regressore con i migliori parametri
    
    
    return best_classificator, best_params
```

```python
rf_classifier_day = RandomForestClassifier()
rf_classifier_nig = RandomForestClassifier()
#Definisco griglia da combinare con gridsearchCV
parameters = {
    'n_estimators': [100, 125, 150],
    'min_samples_split': [2, 3, 4],
    'min_samples_leaf': [2,3,4],
    'max_depth': [7, 10, 12],
    #'min_impurity_decrease': [0.0, 0.1, 0.01] lo commento perchè mi ha sempre dato 0.0, è solo perdita di tempo
}


#Giorno
best_classifier_day, best_params_day = train_classification_model(rf_classifier_day, parameters, 5, x_train_day, y_train_day)

#Notte
best_classifier_nig, best_params_nig = train_classification_model(rf_classifier_nig, parameters, 5, x_train_nig, y_train_nig)
```

```python
#Aggiungo ai best_classfier parametri non specificati nella funzione
best_classifier_day = RandomForestClassifier(**best_params_day, criterion='entropy', oob_score=True, class_weight='balanced', 
                                         bootstrap=True)

best_classifier_nig = RandomForestClassifier(**best_params_nig, criterion='entropy', oob_score=True, class_weight='balanced', 
                                         bootstrap=True)
```

```python
#Addestramento del modello con i migliori parametri
best_classifier_day.fit(x_train_day, y_train_day)
best_classifier_nig.fit(x_train_nig, y_train_nig)
```

```python
# Calcola le probabilità delle classi positive
rf_scores_day = rf_tr_val_day.predict_proba(x_test_day)[:, 1] #predict_proba restituisce un array bidimensionale contenente
#le probabilità stimate per ciascuna riga del set di test, l'1 serve per avere solo le classi positive perchè la prima colonna 
#ha le probabilità di classificare come 0 mentre la seconda ha le probabilità di classificare 1 la stessa riga (infatti la
#somma tra le due probabilità fa 1)
#Notte
rf_scores_nig = rf_tr_val_nig.predict_proba(x_test_nig)[:, 1]
```

```python
# Calcola la curva ROC
fpr_rf_day, tpr_rf_day, thresholds_rf_day = roc_curve(y_test_day, rf_scores_day) #la threshold è la soglia posta alla quale
#l'algoritmo sopra a questo valoredelle probabilità delle y assegna la classe positiva mentre sotto la classe negativa
roc_auc_rf_day = auc(fpr_rf_day, tpr_rf_day)

#Notte
fpr_rf_nig, tpr_rf_nig, thresholds_rf_nig = roc_curve(y_test_nig, rf_scores_nig)
roc_auc_rf_nig = auc(fpr_rf_nig, tpr_rf_nig)
```

```python
#Voglio il valore di threshold migliore quindi quello che per il quale la ROC è più vicina al punto di coord (1,0)

distances_rf_day = ((1 - tpr_rf_day)**2 + fpr_rf_day**2)**0.5 #Calcolo la distanza tra i punti della curva ROC e il punto (0, 1)

best_threshold_idx_rf_day = distances_rf_day.argmin() #Trovo l'indice della threshold con la distanza minima
best_threshold_rf_day = thresholds_rf_day[best_threshold_idx_rf_day]

#Notte
distances_rf_nig = ((1 - tpr_rf_nig)**2 + fpr_rf_nig**2)**0.5
best_threshold_idx_rf_nig = distances_rf_nig.argmin()
best_threshold_rf_nig = thresholds_rf_nig[best_threshold_idx_rf_nig]
```

## ADABoost

Piccola introduzione a come funziona e su cosa si basa ADABoost: ADABoost (Adaptive Boosting) è un algoritmo di apprendimento automatico basato sull'idea di addestrare una sequenza di classificatori deboli per costruire un classificatore forte. Il concetto principale di ADABoost è quello di assegnare un peso ad ogni istanza nel set di addestramento, in modo che gli esempi difficili da classificare abbiano un peso maggiore. L'algoritmo funziona in diversi passaggi iterativi:

- **Inizializzazione dei pesi**: Inizialmente, tutti gli esempi di addestramento hanno lo stesso peso.
- **Addestramento dei classificatori deboli**: Viene addestrato un classificatore debole (ad esempio, un semplice albero decisionale) utilizzando il set di addestramento ponderato. Il classificatore debolmente addestrato viene scelto in base alla sua capacità di minimizzare l'errore pesato sui pesi degli esempi di addestramento.
- **Calcolo dell'errore**: Viene calcolato l'errore del classificatore debolmente addestrato utilizzando i pesi degli esempi di addestramento. Gli esempi classificati erroneamente ricevono un peso maggiore per il passaggio successivo.
- **Aggiornamento dei pesi**: I pesi degli esempi di addestramento vengono aggiornati in base all'errore del classificatore debolmente addestrato. Gli esempi classificati erroneamente ricevono un peso maggiore, mentre gli esempi classificati correttamente ricevono un peso minore.
- **Aggiornamento del classificatore forte**: Il classificatore forte viene costruito combinando i classificatori deboli addestrati nei passaggi precedenti. Il peso di ciascun classificatore debolmente addestrato dipende dalla sua capacità di classificare correttamente gli esempi di addestramento.
- **Ripetizione dei passaggi**: I passaggi 2-5 vengono ripetuti per un numero prefissato di volte o fino a quando non viene raggiunta una condizione di stop.
- **Classificazione finale**: Il classificatore forte addestrato viene utilizzato per classificare nuovi esempi sulla base della votazione dei classificatori deboli.

L'idea principale di ADABoost è che i classificatori deboli, anche se hanno una capacità di classificazione limitata, possono essere combinati in modo da ottenere un classificatore forte che supera le loro limitazioni individuali. Il peso attribuito a ciascun classificatore debolmente addestrato nel classificatore finale dipende dalla sua capacità di classificazione, misurata dall'errore ponderato commesso sui dati di addestramento.

In sintesi, ADABoost è un algoritmo di boosting che costruisce un classificatore forte combinando classificatori deboli in base ai pesi degli esempi di addestramento. Questo approccio adattivo consente di affrontare i problemi di classificazione complessi e di migliorare le prestazioni del modello finale.

```python
#Creo una nuova colonna dentro df_grid in cui clasifico "a parole" l'esattezza o gli errori delle previsioni dei modelli

#ADA

#Giorno
df_grid_square_ada_day['Category'] = 'Unknown'

df_grid_square_ada_day.loc[(df_grid_square_ada_day['HIGH_POW'] > 0) & (df_grid_square_ada_day['PRED_ADA'] > 0), 'Category'] = 'Real Positive'
df_grid_square_ada_day.loc[(df_grid_square_ada_day['HIGH_POW'] > 0) & (df_grid_square_ada_day['PRED_ADA'] == 0), 'Category'] = 'False Negative'
df_grid_square_ada_day.loc[(df_grid_square_ada_day['HIGH_POW'] == 0) & (df_grid_square_ada_day['PRED_ADA'] > 0), 'Category'] = 'False Positive'
df_grid_square_ada_day.loc[(df_grid_square_ada_day['HIGH_POW'] == 0) & (df_grid_square_ada_day['PRED_ADA'] == 0), 'Category'] = 'Real Negative'

df_grid_square_ada_day.dropna(inplace=True) #cancello tutti i NaN perchè sono relativi a squareid non presi in considerazione nella 
#parte urbana di Trento e quindi non sono stati classificati

#Notte
df_grid_square_ada_nig['Category'] = 'Unknown'

df_grid_square_ada_nig.loc[(df_grid_square_ada_nig['HIGH_POW'] > 0) & (df_grid_square_ada_nig['PRED_ADA'] > 0), 'Category'] = 'Real Positive'
df_grid_square_ada_nig.loc[(df_grid_square_ada_nig['HIGH_POW'] > 0) & (df_grid_square_ada_nig['PRED_ADA'] == 0), 'Category'] = 'False Negative'
df_grid_square_ada_nig.loc[(df_grid_square_ada_nig['HIGH_POW'] == 0) & (df_grid_square_ada_nig['PRED_ADA'] > 0), 'Category'] = 'False Positive'
df_grid_square_ada_nig.loc[(df_grid_square_ada_nig['HIGH_POW'] == 0) & (df_grid_square_ada_nig['PRED_ADA'] == 0), 'Category'] = 'Real Negative'

df_grid_square_ada_nig.dropna(inplace=True)

#RF

df_grid_square_rf_day['Category'] = 'Unknown'

df_grid_square_rf_day.loc[(df_grid_square_rf_day['HIGH_POW'] > 0) & (df_grid_square_rf_day['PRED_RF'] > 0), 'Category'] = 'Real Positive'
df_grid_square_rf_day.loc[(df_grid_square_rf_day['HIGH_POW'] > 0) & (df_grid_square_rf_day['PRED_RF'] == 0), 'Category'] = 'False Negative'
df_grid_square_rf_day.loc[(df_grid_square_rf_day['HIGH_POW'] == 0) & (df_grid_square_rf_day['PRED_RF'] > 0), 'Category'] = 'False Positive'
df_grid_square_rf_day.loc[(df_grid_square_rf_day['HIGH_POW'] == 0) & (df_grid_square_rf_day['PRED_RF'] == 0), 'Category'] = 'Real Negative'

df_grid_square_rf_day.dropna(inplace=True) #Cancello per stesso motivo di sopra

#Notte
df_grid_square_rf_nig['Category'] = 'Unknown'

df_grid_square_rf_nig.loc[(df_grid_square_rf_nig['HIGH_POW'] > 0) & (df_grid_square_rf_nig['PRED_RF'] > 0), 'Category'] = 'Real Positive'
df_grid_square_rf_nig.loc[(df_grid_square_rf_nig['HIGH_POW'] > 0) & (df_grid_square_rf_nig['PRED_RF'] == 0), 'Category'] = 'False Negative'
df_grid_square_rf_nig.loc[(df_grid_square_rf_nig['HIGH_POW'] == 0) & (df_grid_square_rf_nig['PRED_RF'] > 0), 'Category'] = 'False Positive'
df_grid_square_rf_nig.loc[(df_grid_square_rf_nig['HIGH_POW'] == 0) & (df_grid_square_rf_nig['PRED_RF'] == 0), 'Category'] = 'Real Negative'

df_grid_square_rf_nig.dropna(inplace=True)
```
